## Data preparation

#### Dropping irrelevant columns

```python
df = df[['coaster_name', 
    #'Length', 'Speed', 
    'Location', 'Status', 
    # 'Opening date',
    #   'Type', 
    'Manufacturer', 
    #'Height restriction', 'Model', 'Height',
    #   'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
    #   'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
    #   'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
    #   'Track layout', 'Fastrack available', 'Soft opening date.1',
    #   'Closing date', 
    'Opened', 
    # 'Replaced by', 'Website',
    #   'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
    #   'Single rider line available', 'Restraint Style',
    #   'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
       # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
       # 'height_value', 'height_unit', 
       'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
df.columns
```

```python
df = df.rename(columns={
    'coaster_name': 'Coaster_Name',
    'year_introduced': 'Year_Introduced',
    'opening_date_clean': 'Opening_Date_Clean',
    'speed_mph': 'Speed_mph',
    'height_ft': 'Height_ft',
    'Inversions_clean': 'Inversions',
    'Gforce_clean': 'Gforce'
})
```

Looking for missing values

```python
df.isna().sum()
```

```python
df.loc[df.duplicated()]
```

```python
df.columns
```

```python
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date_Clean'])].reset_index(drop=True).copy()
```

```python
df
```

## Visualize data

```python
df['Year_Introduced'].value_counts()
```

```python
ax = df['Year_Introduced'].value_counts().head(10).plot(
    kind='bar',
    title='Top Years Coasters Introduced'
    )

ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
```

```python
ax = df['Speed_mph'].plot(kind='hist', bins=20, title='Coaster speed (mph)')

ax.set_xlabel('Speed (mph)')
ax.set_ylabel('Frequency')
```

```python
ax = df['Speed_mph'].plot(kind='kde', title='Coaster speed (mph)')

ax.set_xlabel('Speed (mph)')
ax.set_ylabel('Frequency')
```

```python
df_corr = df[
    ['Year_Introduced', 
     'Speed_mph',
     'Height_ft', 
     'Inversions', 
     'Gforce']
].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr, annot=True)
```

## What are the locations with the fastest roller coasters (minimum of 10)

```python
df.query('Location != "Other"').groupby('Location')['Speed_mph'].agg(['mean', 'count']).query('count >= 10').sort_values('mean')['mean'].plot(kind='barh', title='Average  Coast Speed by Location')
```
